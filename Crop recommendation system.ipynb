```python
# source code for recommendation system
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv("C:/Users/UltraBook 3.1/Desktop/data_analysis projects/crop_recommendation_system/Crop_recommendation.csv")
```

```python
data.label.value_counts()
```

```python
# no missing values
data.isnull().sum()
```

```python
X = data.drop(columns={'label'})
```

```python
# check for any correlations between the features
corr_matrix = X.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# from the heat map no strong relationships are detected hence proceed
# get the dependent variable 
Y = data['label']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.25, shuffle = True, random_state = 0)
```

```python
# initialize gaussian naive bayes model
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

```python
# apply to test data
y_predict = gnb.predict(X_test)
```

```python
accuracy_value = accuracy_score(Y_test, y_predict)
```

```python
print('accuracy-score:', accuracy_value)
```

```python
cm = confusion_matrix(Y_test, y_predict)
```

```python
sns.heatmap(cm, annot= True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues')
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Confusion Matrix - score:'+str(accuracy_score(Y_test, y_predict))
plt.title(all_sample_title, size = 15)
```
